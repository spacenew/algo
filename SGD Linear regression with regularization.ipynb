```python
class SGDLinearRegression:
    
    # Создаем инициализатор класса
    def __init__(self, step_size, epochs, batch_size, reg_weight):
        
        # step_size - величина шага градиентного спуска
        self.step_size = step_size
       
        # epochs - 1 эпоха  - проход по всем батчам
        self.epochs = epochs
        
        # Для уменьшения времени работы применим алгоритм к небольшим частям обучающей выборки (batch)
        self.batch_size = batch_size
        
        # reg_weight - вес регуляризации
        self.reg_weight = reg_weight
    
    #Обучаем модель
    def fit(self, train_features, train_target):
        
        X = np.concatenate((np.ones((train_features.shape[0], 1)), train_features), axis=1)        
        y = train_target
        
        # Задаем начальные веса = 0
        w = np.zeros(X.shape[1])
        
        for _ in range(self.epochs):
            # Вычисляем количество батчей
            batches_count = X.shape[0] // self.batch_size
            for i in range(batches_count):
                # Находим начало батча с номером i - индексом первого элемента
                begin = i * self.batch_size
                # Находим конец батча с номером i  - (начало следующего батча)
                end = (i + 1) * self.batch_size
                
                X_batch = X[begin:end, :]
                y_batch = y[begin:end]
                
                # Находим градиент по батчу
                gradient = 2 * X_batch.T.dot(X_batch.dot(w) - y_batch) / X_batch.shape[0]
                
                # копируем вектор w, чтобы его не менять
                reg = 2 * w.copy()
                
                # Приравняем элемент с индексом ноль к нулю, т.к обычно сдвиг не включается в регуляризацию
                reg[0] = 0
                
                # К градиенту функции добавляем слагаемое для регулязации
                gradient += self.reg_weight * reg
                
                # Делаем шаг по антиградиенту
                w -= self.step_size * gradient

        self.w = w[1:]
        self.w0 = w[0]
    
    # Предсказываем значение метрикиметрику
    def predict(self, test_features):
        # Рассчитываем значение функции потерь R2
        return test_features.dot(self.w) + self.w0
```
